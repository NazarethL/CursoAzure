La Release 4.0 de la Calculadora, deberá almacenar en la(s) Estructura(s) de Almacenamiento en Memoria, el histórico de las operaciones realizadas; hasta que el usuario finaliza la ejecución.

Al finalizar la ejecución, se debe ofrecer al usuario la opción de “mostrar” por pantalla, dicho histórico.

```python
def datos():
    try:
        operando1= float(input("Introduce el primer operando:"))
        operando2= float(input("Introduce el segundo operando:"))
    
        operador= input('''¿Qué operación quiere realizar?
        1 - Sumar
        2 - Restar
        3 - Multiplicar
        4 - Dividir
        5 - Salir
        ''')
        return operando1, operando2, int(operador)
    except ValueError:
        print("Introduzca un número válido.")
        return 'error', 0, 0

def suma(a, b):
    solucion = a + b
    return solucion
    
def resta(a, b):
    solucion = a - b
    return solucion

def multiplicacion(a, b):
    solucion = a * b
    return solucion

def division(a, b):
    solucion = a / b
    return solucion

o = 0
i = 0
historial = {}

while o != 5:

    a, b, o = datos()
    
    if o==1:
        historial[i]=[a,'+', b, suma(a,b)]
        i = i+1
        print ("Solucion: {0}".format(suma(a,b)))
    elif o==2:
        historial[i]=[a,'-', b, resta(a,b)]
        i = i+1
        print("Solucion: {0}".format(resta(a,b)))
    elif o==3:
        historial[i]=[a,'*', b, multiplicacion(a,b)]
        i = i+1
        print("Solucion: {0}".format(multiplicacion(a,b)))
    elif o==4:
        historial[i]=[a,'/', b, division(a,b)]
        i = i+1
        print("Solucion: {0}".format(division(a,b)))
    elif o==5:
        break
    elif a!='error':
        print("No ha introducido una operación válida, vuelva a intentarlo.")
        
mostrar = input("¿Desea mostrar el historial? S/N")

if mostrar=='S':
    print(historial)
elif mostrar=='N':
    print("Hasta la próxima")
else:
    print("Se considerará un no. Hasta la próxima.")
```
